```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfc = pd.read_stata("p50_constant.dta")
dfp = pd.read_stata("p50_positive.dta")
dfn = pd.read_stata("p50_negative.dta")
dfo = pd.read_stata("p50_observed.dta")
```

```python
constante = dfc.groupby(["year"])["survivalB2p50"].sum()
positiva = dfp.groupby(["year"])["survivalB2p50"].sum()
negativa = dfn.groupby(["year"])["survivalB2p50"].sum()
observado = dfo.groupby(["year"])["survivalB2p50"].sum()
```

```python
x = negativa.index
observador = [4, 3, 3, 6, 23, 39, 29, 47, 59, 37, 47, 49, 91, 75, 69, 55, 26, 25, 38, 20, 38, 39, 24, 17, 20, 24, 27, 10, 7, 8, 5, 2, 1, 2, 1, 1]
alive = [4, 7, 7, 12, 31, 60, 70, 100, 126, 126, 138, 166, 205, 220, 218, 207,193, 166, 157, 139, 138, 143, 124, 119, 127, 138, 138, 109, 74, 55, 46, 38, 36, 29, 27, 27]

# Create the plot
plt.figure(figsize=(10, 6))

# Plotting each function with labels for the legend
plt.plot(x, constante, label='Constant', color='blue')
plt.plot(x, positiva, label='Positive', color='green')
plt.plot(x, negativa, label='Negative', color='red')
plt.plot(x, observado, label='Observed', color='black')
# plt.plot(x, alive, label='Alive', color='purple')
plt.legend(loc='upper left')

# Creating a second y-axis on the right
ax2 = plt.gca().twinx()

# # Plotting the 'alive' function on the secondary y-axis
ax2.plot(x, alive, label='Alive (right axis)', color='purple')

# Adding title and labels
# plt.title('Comparacin de Funciones: Constante, Positiva y Negativa')
# plt.xlabel('Eje X')
# plt.ylabel('Eje Y')

# Adding a legend
# Adding legends for both axes

ax2.legend(loc='upper right')

# Show the plot
# plt.grid(True)

plt.savefig("entrada.jpeg", dpi = 500)
plt.show()
```
